```python
data = pd.DataFrame(np.random.randint(1,20000,(1000,2)), columns = ['Celsius','KM'])
data
```

```python
data['Fehren'] = data['Celsius'] * ( 9.0 / 5.0 ) + 32
data
```

```python
data['Inch'] = data['KM'] * 39370.0787 # (KM * 1000 * 100 / 2.54)
data
```

```python
new_row = [30,5]
new_row.append(new_row[0] *  (9.0 / 5.0 ) + 32 )
new_row.append(new_row[1] * 39370.0787)
new_row
```

```python
data.loc[data.shape[0]] = new_row
data
```

```python
data[data['KM'] > 500]
```

```python
data[(data['KM'] > 500) & (data['Celsius'] != 300)]
```

```python
data[(data['KM'] > 500) & (data['Celsius'] == 300)]
```

```python
data[(data['KM'] > 500) & ~(data['Celsius'] == 300)]
```

```python
data[(data['KM'] > 500) | ~(data['Celsius'] == 300)]
```

```python
data[data['KM'].between(700,750)]
```

```python
data[data['KM'].isin([715,721])]
```
